```python
b, a = signal.butter(3, 0.1, 'low', analog=False)
#w, h = signal.freqs(b, a)

w, h = signal.freqz(b, a)

#plt.plot(w, 20 * np.log10(abs(h)))
plt.subplot(4, 1, 1)
plt.plot(w/np.pi, 20 * np.abs(h))
plt.xscale('log')
plt.title('Butterworth lowpass filter mag response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(1, color='green') # cutoff frequency
plt.show()

b, a = signal.butter(3, 0.1, 'high', analog=False)
w, h = signal.freqz(b, a)

plt.subplot(4, 1, 2)
#plt.plot(w, 20 * np.log10(abs(h)))
plt.plot(w/np.pi, 20 * np.abs(h))
plt.xscale('log')
plt.title('Butterworth highpass filter mag response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(1, color='green') # cutoff frequency
plt.show()

b, a = signal.butter(3, [0.01, 0.1], 'bandpass', analog=False)
#w, h = signal.freqs(b, a)

w, h = signal.freqz(b, a)
plt.subplot(4, 1, 3)
#plt.plot(w, 20 * np.log10(abs(h)))
plt.plot(w/np.pi, 20 * np.abs(h))
plt.xscale('log')
plt.title('Butterworth bandpass filter mag response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(1, color='green') # cutoff frequency
plt.show()

b, a = signal.butter(3, [0.01, 0.1], 'bandstop', analog=False)
#w, h = signal.freqs(b, a)

w, h = signal.freqz(b, a)
plt.subplot(4, 1, 4)
#plt.plot(w, 20 * np.log10(abs(h)))
plt.plot(w/np.pi, 20 * np.abs(h))
plt.xscale('log')
plt.title('Butterworth bandstop filter mag response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(1, color='green') # cutoff frequency
plt.show()
```

```python
def dirac(n=0, N=40):
    # Check that the parameter are valid
    if n > N - 1:
        raise ValueError('n should be inferior to N-1')
    
    # Create the array
    s = np.zeros((N, ))
    # Place the dirac
    s[n] = 1
    
    return s


plt.stem(dirac(20))
plt.xlabel(r'$n$')
plt.ylabel(r'$\delta(k-n)$')
plt.title('Dirac function')


```
